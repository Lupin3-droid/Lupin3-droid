```python
#import dependancies.
import pandas as pd
import numpy as np
```

```python
#import data
df=pd.read_excel("/content/kazi_m.xlsx")
df.head()
```

```python
#checking for missing values.
df.isnull().sum()
```

```python
#checking for duplicates based on 'Mentee_Name' and 'Session_Date' columns
duplicates = df[df.duplicated(subset=["Mentee_Name","Session_Date"], keep=False)]
# Displaying the duplicates
print(duplicates)
# Counting the number of duplicate rows
num_duplicates = duplicates.shape[0]
print(f"Number of duplicate rows: {num_duplicates}")
```

```python
#dealing with missing values
df2=df.drop(columns=["Points_Awarded"])
df2.dropna(inplace=True)
df2.info()
df2.head(14)
```

```python
#checking for duplicates based on 'Mentee_Name' and 'Session_Date' columns
duplicates = df2[df2.duplicated(subset=["Mentee_Name","Session_Date"], keep=False)]
# Displaying the duplicates
print(duplicates)
# Counting the number of duplicate rows
num_duplicates = duplicates.shape[0]
print(f"Number of duplicate rows: {num_duplicates}")
```

```python
#dealing with duplicates in df2
df2.drop_duplicates(subset=["Mentee_Name","Session_Date"], keep="first", inplace=True)
df2.head(14)
df2.duplicated().sum()
```
